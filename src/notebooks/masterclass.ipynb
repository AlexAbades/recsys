```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Given data
weights = np.array([60, 60, 62, 64, 65, 68, 70, 72, 74, 76, 78, 80, 82, 84, 86, 88, 90, 70, 72, 71, 73, 75, 77, 70])
heights = np.array([1.55, 1.60, 1.60, 1.63, 1.69, 1.70, 1.73, 1.76, 1.77, 1.79, 1.8, 1.75, 1.85, 1.80, 1.87, 1.92, 1.88, 1.69, 1.74, 1.65, 1.76, 1.67, 1.78, 1.80])

# Stack weights and heights into a single matrix
data = np.vstack([weights, heights]).T

# Perform PCA
pca = PCA(n_components=2)
pca.fit(data)

# Compute the centroid (mean of the data)
centroid = np.mean(data, axis=0)

# Determine the principal axis direction (first principal component)
direction = pca.components_[0]

# Decide on the length of the line
line_length = 30  # Adjust as needed

# Calculate the endpoints of the line
point1 = centroid - direction * (line_length / 2)
point2 = centroid + direction * (line_length / 2)

# Plotting
plt.figure(figsize=(8, 5))
plt.scatter(weights, heights, color='dodgerblue', edgecolor='black', linewidth=1.5, s=100)
plt.plot([point1[0], point2[0]], [point1[1], point2[1]], '--k', linewidth=2)
plt.title('Weights vs. Heights ', fontsize=16, fontweight='bold')
plt.xlabel('Weight (kg)', fontsize=14)
plt.ylabel('Height (m)', fontsize=14)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a range of positive values (excluding 0 since log(0) is undefined)
x_values = np.linspace(0.1, 10, 400)
y_values = np.log(x_values)  # Natural logarithm (base e)

# Plotting the logarithmic function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='y = log(x)')
plt.title('Natural Logarithm Function')
plt.xlabel('x')
plt.ylabel('log(x)')
plt.legend()
plt.grid(True)
plt.show()
```
